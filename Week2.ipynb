Membuat bentuk Geometri 2D dari OpenCV. Diawali dengan pembuatan black image dengan tipe data int16.

```python
black_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
plt.imshow(black_img)
```

Kemudian menambahkan bentuk persegi panjang sesuai koordinat pt1 dan pt2

```python
# perhatikan koordinat titik2 pt1 dan pt2
cv.rectangle(black_img, pt1=(384, 10), pt2=(510, 150), color=(0, 255, 0), thickness=10)
plt.imshow(black_img)
```

Selanjutnya ditambah menambahkan bentuk persegi sesuai koordinat pt1 dan pt2 yang tertulis pada kode program.

```python
cv.rectangle(black_img, pt1=(200, 200), pt2=(300, 300), color=(0, 0, 255), thickness=15)
plt.imshow(black_img)
```

Tahap selanjutnya ditambah menambahkan bentuk lingkaran sesuai radius yang tertulis pada kode program.

```python
cv.circle(black_img, center=(100, 100), radius=50, color=(255, 0, 0), thickness=8)
plt.imshow(black_img)
```

Kemudian dilakukan penambahan garis sesuai koordinat pt1 dan pt2 sebagai berikut.

```python
cv.line(black_img, pt1=(0, 0), pt2=(512, 512), color=(255, 255, 255), thickness=5)
plt.imshow(black_img)
```

Penambahan text dengan font yang telah tertulis dengan ukuran yang sudah ditentukan

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='HELLO', org=(10, 500), fontFace=font, fontScale=4, color=(255, 255, 255), thickness=2, lineType=cv.LINE_AA)
plt.imshow(black_img)
```

Pembuatan black image kembali dilakukan dengan tipe data int32

```python
black_img2=np.zeros(shape=(512, 512, 3), dtype=np.int32)
plt.imshow(black_img2)
```

Berikut adalah kode program untuk inisialisasi NumPy array dengan tipe data int32

```python
vertices = np.array([[100,300], [200,200],[400,300], [200,400]], dtype=np.int32)
vertices
```

Array tersebut kemudian di reshape sebagai berikut

```python
pts = vertices.reshape((-1, 1, 2)) # nilai 2 untuk menunjukkan bahwa tiap titik dibuat 3 channel yang mewakili R, G, dan B
pts
```

Penambahan polyline pada black image kedua yang telah dibuat.

```python
cv.polylines(black_img2, [pts], isClosed=True, color=(255, 0, 0), thickness=5)
plt.imshow(black_img2)
```

5. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!

```python
import random

img = cv.imread('/content/drive/MyDrive/PCVK/kitten01.jpg')

height, width, _ = img.shape

colors = [
    (255, 0, 0),    # Biru
    (0, 255, 0),    # Hijau
    (0, 0, 255),    # Merah
    (255, 255, 0),  # Cyan
    (0, 255, 255)   # Kuning
]

img_boxes = img.copy()

for i in range(5):
    x1, y1 = random.randint(0, width - 50), random.randint(0, height - 50)
    x2, y2 = x1 + random.randint(850, 850), y1 + random.randint(850, 850)
    color = random.choice(colors)
    cv.rectangle(img_boxes, (x1, y1), (x2, y2), color, 100)

plt.imshow(cv.cvtColor(img_boxes, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Gambar dengan 5 Kotak Outline")
plt.show()
```

6. Tampilkan image dengan posisi terbalik!

```python
flipped = cv.flip(img_boxes, 1)
plt.imshow(cv.cvtColor(flipped, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Terbalik")
```
