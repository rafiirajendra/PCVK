Akses file image lokal (KTM) dan tampilkan menggunakan OpenCV dan
Matplotlib

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
img = cv.imread('/content/drive/MyDrive/PCVK/lisa_ktm.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

bentuknya.

### Kelompok 7 bagian Malang, bulan tahun dan NIP Direktur (karena tidak ada NIP jadinya NIM)

```python
edited = img.copy()
# 1. Tutup tulisan "Malang" (pakai rectangle biru)
cv.rectangle(edited, (350, 100), (500, 150), (0, 0, 255), -1)

# 2. Tutup bulan/tahun "02/26" (pakai ellipse hijau)
cv.ellipse(edited, (930, 650), (100, 50), 0, 0, 360, (0, 255, 0), -1)

# 3. Tutup NIM "2341721234" (pakai rectangle kuning)
cv.rectangle(edited, (380, 900), (950, 850), (255, 255, 0), -1)

# Tampilkan hasil
plt.figure(figsize=(10,6))
plt.imshow(edited)
plt.title('Kelompok 7 bagian Malang, bulan tahun dan NIP Direktur (karena tidak ada NIP jadinya NIM)')
plt.axis('off')
plt.show()
```
